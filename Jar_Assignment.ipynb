Regional Performance Insights:

```python
Top_states=df_final['State'].value_counts().head(5).index.tolist()
states_df=df_final[df_final['State'].isin(Top_states)]
```

```python
states_df
```

```python
Top_states
```

```python
State_data=states_df.groupby('State').agg({'Amount':'sum', 'Profit':'mean', 'Order ID':'count'}).rename(columns={'Order ID':'Order Count'}).reset_index()
```

```python
State_data
```

```python
State_data = State_data.sort_values(by='Amount', ascending=False).reset_index(drop=True)
```
